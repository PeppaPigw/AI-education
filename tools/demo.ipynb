```python
# æ±‡æ€»é¢˜ç›®
import os
from docx import Document

def merge_docx_to_txt(directory_path, output_filename):
    all_text = []

    for filename in os.listdir(directory_path):
        if filename.endswith(".docx") and not filename.startswith("~"):
            file_path = os.path.join(directory_path, filename)
            
            try:
                # 1. è¯»å– .docx æ–‡ä»¶
                document = Document(file_path)
                
                # 2. æå–æ–‡æœ¬å†…å®¹
                file_text = []
                for para in document.paragraphs:
                    file_text.append(para.text)
                all_text.append("\n".join(file_text))                
                print(f"âœ… å·²æˆåŠŸå¤„ç†æ–‡ä»¶: {filename}")
                
            except Exception as e:
                print(f"âŒ å¤„ç†æ–‡ä»¶ {filename} æ—¶å‡ºé”™: {e}")

    # 4. å°†æ‰€æœ‰å†…å®¹å†™å…¥ .txt æ–‡ä»¶
    output_path = os.path.join(directory_path, output_filename)
    try:
        # ä½¿ç”¨ 'w' æ¨¡å¼å†™å…¥ï¼Œå¦‚æœæœ‰åŒåæ–‡ä»¶ä¼šè¦†ç›–
        with open(output_path, 'w', encoding='utf-8') as outfile:
            outfile.write("".join(all_text))
        
        print("\nğŸ‰ æ‰€æœ‰æ–‡ä»¶å†…å®¹å·²æˆåŠŸåˆå¹¶åˆ°ï¼š")
        print(output_path)
        
    except Exception as e:
        print(f"âŒ å†™å…¥è¾“å‡ºæ–‡ä»¶æ—¶å‡ºé”™: {e}")
for i in range(1, 7):
    target_directory = f"./{i}/"
    output_file = f"Q{i}.txt"
    merge_docx_to_txt(target_directory, output_file)
```
